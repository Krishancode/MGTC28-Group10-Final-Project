```python

sns.lineplot(x='Fiscal Week', y='Need State 2', data=dfTRD, hue='Fiscal Year', palette='viridis')


dfTRD['Week Range'] = pd.cut(dfTRD['Fiscal Week'], bins=[1, 10, 20, 30, 40, 50], labels=['1-10', '11-20', '21-30', '31-40', '41-50'])

# Now, you can group by the 'Week Range' column and perform aggregations as needed
grouped_data = dfTRD.groupby('Week Range').agg({'Fiscal Year': 'sum', 'Fiscal Week': 'mean'})

# Display the grouped data
print(grouped_data)

plt.title('Total Sales vs. Total Trade Spend')
plt.xlabel('Total Trade Spend')
plt.ylabel('Sales')
plt.legend()
plt.show()

```

```python
# Total Sales Compared to Sales Trend for Need Sate 1

dfTRD['Week Range'] = pd.cut(dfTRD['Fiscal Week'], bins=[1, 10, 20, 30, 40, 50], labels=['1-10', '11-20', '21-30', '31-40', '41-50'])

# Group by the 'Week Range' column
grouped_data = dfTRD.groupby(['Week Range', 'Fiscal Year']).agg({'Need State 1': 'sum'}).reset_index()

# Graph the grouped data into a Bar plot
sns.barplot(x='Week Range', y='Need State 1', data=grouped_data, hue='Fiscal Year', palette='viridis')

# Set axis titles and labels
plt.title('Total Sales vs. Total Trade Spend for Need State 1')
plt.xlabel('Week Range')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
